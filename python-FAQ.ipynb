```python
a = ['Apple', ['Mango','Green'], 'Banana']
b = a.copy()

a.remove('Banana')
b[1][1] = 'Yellow'
print(a)
print(b)
```

```python
for i in range(5,0,-1):
    for j in range(5):
        if j < i:
            print(j + 1, end='\t')
    print("\n")
```

```python
r = 5
for i in range(r):
    if j in range(r*2):
        if j <= (r + i) or j >= (r - i):
            print(i, j, "*", end=' ')
        else:
            print(i, j, " ", end=' ')
    print("\n")
```
