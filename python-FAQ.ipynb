```python
for i in range(r):
    s = 0
    if i % 2 == 0:
        s = 1
    for j in range(r):
        if j <= i:
            print(s, end='   ')
            s = 1 - s
    print("\n")
```

```python
s = 2 * (r - 2)
for i in range(r-1):
    for j in range(i + 1):
        print(j+1, end='   ')
    for j in range(s):
        print(" ", end='   ')
    for j in range(i, -1, -1):
        print(j+1, end='   ')
    s -= 2
    print("\n")
```

```python
for i in range(r):
    for j in range(r - i - 1):
        print(" ", end='   ')
    
    ch = 65
    bp = (2 * i + 1) / 2 - 1
    for j in range(2 * i + 1):
        print(chr(ch), end='   ')
        if j <= bp:
            ch += 1
        else:
            ch -= 1
    
    for j in range(r - i - 1):
        print(" ", end='   ')
    print("\n")
```

```python
for i in range(r, 0, -1):
    for j in range(i, r + 1):
        print(chr(ord('A') + j - 1), end='   ')
    print("\n")

# OR
for i in range(r):
    ch = ord('A') + r - 1 - i # 65 + 5 - 1 - i
    for j in range(i+1):
        print(chr(ch), end='   ')
        ch += 1
    print("\n")
```

```python
for i in range(r):
    for j in range(r - i):
        print("*", end='   ')
    for j in range(2 * i):
        print(" ", end='   ')
    for j in range(r - i):
        print("*", end='   ')
    print("\n")
for i in range(r,0,-1):
    for j in range(r - i + 1):
        print("*", end='   ')
    for j in range(2 * i - 2):
        print(" ", end='   ')
    for j in range(r - i + 1):
        print("*", end='   ')
    print("\n")
```

```python
sp = 2 * r - 2
for i in range(2 * r - 1):
    s = i + 1
    if i > r - 1:
        s = 2 * r - i - 1
    for j in range(s):
        print("*", end='   ')
    for j in range(sp):
        print(" ", end='   ')
    for j in range(s):
        print("*", end='   ')
    if i < r - 1:
        sp -= 2
    else:
        sp += 2
    print("\n")
```

```python
for i in range(r):
    for j in range(r):
        if i == 0 or i == r-1 or j == 0 or j == r-1:
            print("*", end='   ')
        else:
            print(" ", end='   ')
    print("\n")
```

```python
for i in range(2 * r - 1):
    for j in range(2 * r - 1):
        tp = i
        lt = j
        bt = (2*r - 2) - i
        rt = (2*r - 2) - j
        print(r - min(tp,lt,bt,rt), end='   ')
    print("\n")
```
